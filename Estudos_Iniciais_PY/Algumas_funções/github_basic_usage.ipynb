# Guia Básico de Uso do GitHub com Códigos

Este notebook fornece uma explicação básica dos principais comandos do Git que você pode usar ao trabalhar com repositórios GitHub. O Git é um sistema de controle de versão distribuído, enquanto o GitHub é uma plataforma de hospedagem de repositórios Git.


## 1. Clonar um Repositório

Para começar a trabalhar em um projeto existente, você precisa clonar o repositório para o seu ambiente local.

```bash
git clone https://github.com/usuario/nome-do-repositorio.git
```

Esse comando cria uma cópia completa do repositório em seu ambiente local.

## 2. Verificar o Status

Depois de fazer alterações em seu repositório local, você pode verificar o status das suas alterações usando:

```bash
git status
```

Este comando mostra quais arquivos foram modificados e quais ainda não foram adicionados ao próximo commit.

## 3. Adicionar Arquivos ao Índice

Para preparar as mudanças para o commit, você precisa adicionar os arquivos modificados ao índice (ou stage).

```bash
git add nome-do-arquivo
```

Ou, para adicionar todos os arquivos modificados:

```bash
git add .
```

## 4. Fazer um Commit

Depois de adicionar os arquivos ao índice, você pode salvar suas mudanças em um commit.

```bash
git commit -m "Mensagem descritiva do commit"
```

A mensagem de commit deve ser clara e descritiva, indicando o que foi alterado.

## 5. Enviar Mudanças para o Repositório Remoto

Finalmente, para compartilhar suas mudanças com os outros colaboradores do projeto, você precisa enviar seus commits para o repositório remoto.

```bash
git push origin nome-da-branch
```

Substitua `nome-da-branch` pela branch na qual você está trabalhando, geralmente `main` ou `master`.

## 6. Puxar Mudanças do Repositório Remoto

Para manter seu repositório local atualizado com as últimas mudanças do repositório remoto, use o comando `pull`.

```bash
git pull origin nome-da-branch
```

Isso irá buscar e mesclar as mudanças mais recentes do repositório remoto na sua branch atual.
