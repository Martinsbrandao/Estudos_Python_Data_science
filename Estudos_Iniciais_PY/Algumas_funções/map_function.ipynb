# Explicação da Função `map` em Python

A função `map` em Python é usada para aplicar uma função a cada item de um iterável (como uma lista, tupla, etc.) e retornar um objeto `map` (um iterador) com os resultados. Em outras palavras, ela mapeia uma função a todos os elementos de uma coleção.


### Sintaxe básica:

```python
map(function, iterable)
```

- `function`: Uma função que você deseja aplicar a cada item do iterável.
- `iterable`: Um iterável (como uma lista, tupla, etc.) cujos elementos serão passados para a função.


### Exemplo:

```python
# Função que será aplicada a cada item
def square(x):
    return x * x

# Lista de números
numbers = [1, 2, 3, 4, 5]

# Usando map para aplicar a função square a cada item da lista
squared_numbers = map(square, numbers)

# Convertendo o resultado para uma lista para visualizar
print(list(squared_numbers))
```


**Saída:**

```python
[1, 4, 9, 16, 25]
```


No exemplo acima, a função `square` é aplicada a cada elemento da lista `numbers`, e os resultados são armazenados no iterador `squared_numbers`, que é convertido para uma lista para exibição.

`map` é particularmente útil quando você deseja aplicar uma operação a todos os elementos de uma coleção de forma eficiente e concisa.
